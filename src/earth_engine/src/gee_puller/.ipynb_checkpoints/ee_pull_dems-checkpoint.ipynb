```python
import ee 
import os
import geopandas as gpd
import ee_helpers
```

```python
ee.Initialize()
ee.Authenticate()
```

```python
read_dir = os.path.join(os.path.expanduser('~'),"Desktop","projects","GlacierView","glims-processor","output")
read_file = "joined_474_bb.shp"
df = gpd.read_file(os.path.join(read_dir,read_file))
```

```python
out_dir = os.path.join("data","joined_dems", "dems")
for row in df.iterrows():
    glims_id = row[1].glac_id
    bounding_box = eval(row[1].bboxes)
    out_dir = os.path.join(out_dir)
    ee_helpers.export_nasa_dems(glims_id, bounding_box, out_dir)
  
```
